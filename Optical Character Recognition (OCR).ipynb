```python
#To perform graphical plot into the context
import matplotlib.pyplot as plt

#To perform the multi-dimensional operation
import numpy as np

#To load an image from the specified file
import cv2 as cv

#Optical Character Recognition Tool 
import pytesseract
```

```python
imgThreshold = cv.Canny(gray,100,0)

backup = imgThreshold.copy()

plt.imshow(imgThreshold)
```

### To perform fundamental morphological operations

```python
kernel = np.ones((5, 5))

#To apply dilation 
imgDial = cv.dilate(imgThreshold, kernel, iterations=2) 

#To apply erosion 
imgThreshold = cv.erode(imgDial, kernel, iterations=1)  
```
